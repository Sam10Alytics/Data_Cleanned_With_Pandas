```python
def remove_white_space(text):
   if type(text) == str:
      return text.strip()
   return text   
```

```python
travels_dataset.info()
```

```python
travels_dataset['Travel Start Date'] = travels_dataset['Travel Start Date'].apply(remove_white_space)
```

```python
travels_dataset.info()
```

```python
travels_dataset['Travele End  Date'] = travels_dataset['Travel End Date'].apply(remove_white_space)
travels_dataset['Travel End Date'] = pd.to_datetime(travels_dataset['Travel End Date'])
```

```python
""" extract state"""

def extract_state(text):
    result = text.split(',')[1]
    return result


""" extract scity"""

def extract_city(text):
    result = text.split(',')[0]
    return result
```

```python
# extract state and city fom destination
travels_dataset['state'] = travels_dataset['Destination'].apply(extract_state)
travels_dataset['city'] = travels_dataset['Destination'].apply(extract_city)
```

```python
def extract_first_name(text):
    '''extract first name'''
    result = text.split(' ')[0]
    return result


def extract_last_name(text):
    '''extract last name'''
    result = text.split(' ')[1]
    return result
```

```python
travels_dataset['first_name'] = travels_dataset['Employee'].apply(extract_first_name)
travels_dataset['last_name'] = travels_dataset['Employee'].apply(extract_last_name)
```

```python
columns = ['first_name','last_name','state','city','Travel Start Date','Travel End Date','Actual Total Expenses','Purpose Of Travel']
final_result = travels_dataset[columns].copy()
```

```python
final_result.to_csv('Cleaned_travels_dataset.csv')
```

```python
def remove_special_characters(name):
    result = name.replace('@','')
    result = result.replace('!','')
    return result
    
```

```python
travels_dataset['Employee'] = travels_dataset['Employee'].apply(remove_special_characters)
```

```python
travels_dataset.to_csv('Cleaned_travels_dataset.csv')
```
